__QUESTION THREE: Searching and Sorting Algorithms__

```python
arr = [21, 10, 45, 32, 67, 5, 18]

# (a) Linear Search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

print("Linear Search (32):", linear_search(arr, 32))

# (b) Binary Search (list must be sorted)
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

sorted_arr = sorted(arr)
print("Binary Search (45):", binary_search(sorted_arr, 45))

# (c) Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print("Bubble Sort:", bubble_sort(arr.copy()))
```

__QUESTION FOUR: Linked Lists Implementation__

```python
# (a) Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# (b) LinkedList Class
class LinkedList:
    def __init__(self):
        self.head = None

    # i. Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # ii. Insert at end
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # iii. Display all elements
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# Example usage
ll = LinkedList()
ll.insert_beginning(10)
ll.insert_end(20)
ll.insert_end(30)
ll.display()
```

__QUESTION FIVE: Application of Data Structures__

A Queue is most suitable.This is because students are served in FIFO (First-In-First-Out) order.

```python
from collections import deque

class FinanceQueue:
    def __init__(self):
        self.queue = deque()

    # i. Add student
    def add_student(self, student):
        self.queue.append(student)
        print(f"{student} added to queue.")

    # ii. Serve student
    def serve_student(self):
        if self.queue:
            served = self.queue.popleft()
            print(f"{served} has been served.")
        else:
            print("No students in queue.")

    # iii. Display queue
    def display_queue(self):
        print("Current Queue:", list(self.queue))

# Example usage
fq = FinanceQueue()
fq.add_student("Alice")
fq.add_student("Bob")
fq.display_queue()
fq.serve_student()
fq.display_queue()
```
